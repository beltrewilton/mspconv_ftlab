```python
import stable_whisper
model = stable_whisper.load_model('base')
```

### Roberta base-go

https://huggingface.co/SamLowe/roberta-base-go_emotions

27 emotions + neutral

```python
from transformers import pipeline

classifier = pipeline(task="text-classification", model="SamLowe/roberta-base-go_emotions", top_k=None)
```

```python
for result in results:
    model_outputs = classifier(result.text)
    print(result.text)
    print(model_outputs[0][0]['label'], round(model_outputs[0][0]['score'],2))
    print(model_outputs[0][1]['label'], round(model_outputs[0][1]['score'],2))
    print(model_outputs[0][2]['label'], round(model_outputs[0][2]['score'],2))
    print('=========================================================')
    print('')
```

### Roberta twitter 

https://huggingface.co/cardiffnlp/twitter-roberta-base-emotion-multilabel-latest

11 emotions

```python
from transformers import pipeline

pipe = pipeline("text-classification", model="cardiffnlp/twitter-roberta-base-emotion-multilabel-latest", return_all_scores=True)
```

```python
for result in results:
    model_outputs = pipe(result.text)
    print(result.text)
    print(model_outputs[0][0]['label'], round(model_outputs[0][0]['score'],2))
    print(model_outputs[0][1]['label'], round(model_outputs[0][1]['score'],2))
    print(model_outputs[0][2]['label'], round(model_outputs[0][2]['score'],2))
    print('=========================================================')
    print('')
```
